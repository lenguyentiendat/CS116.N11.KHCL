```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_palette("Set3")
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import os
```

#Read dataset

```python
df = pd.read_csv('/content/drive/MyDrive/Covid Data.csv')
```

```python
df.head()
```

```python
df.info()
```

#Data preprocessing

```python
df = df[(df.PNEUMONIA == 1) | (df.PNEUMONIA == 2)]
df = df[(df.DIABETES == 1) | (df.DIABETES == 2)]
df = df[(df.COPD == 1) | (df.COPD == 2)]
df = df[(df.ASTHMA == 1) | (df.ASTHMA == 2)]
df = df[(df.INMSUPR == 1) | (df.INMSUPR == 2)]
df = df[(df.HIPERTENSION == 1) | (df.HIPERTENSION == 2)]
df = df[(df.OTHER_DISEASE == 1) | (df.OTHER_DISEASE == 2)]
df = df[(df.CARDIOVASCULAR == 1) | (df.CARDIOVASCULAR == 2)]
df = df[(df.OBESITY == 1) | (df.OBESITY == 2)]
df = df[(df.RENAL_CHRONIC == 1) | (df.RENAL_CHRONIC == 2)]
df = df[(df.TOBACCO == 1) | (df.TOBACCO == 2)]
```

```python
df.head()
```

```python
#Convert null data to 2 in pregnant
df.PREGNANT = df.PREGNANT.replace(98, 2)
df.PREGNANT = df.PREGNANT.replace(97, 2)


#INTUBED Variable has so many null values. We delete it
df.drop("INTUBED", axis = 1, inplace = True)
df.drop("ICU", axis = 1, inplace = True)
df.head()
```

```python
df.CLASIFFICATION_FINAL.value_counts()
```

```python
#1,2,3 means you have covid, 4,5,6,7 means not, so i convert 1,2,3 to 1 and 4,5,6,7 to 2.
df.CLASIFFICATION_FINAL = df.CLASIFFICATION_FINAL.replace([1,2,3], 1)
df.CLASIFFICATION_FINAL = df.CLASIFFICATION_FINAL.replace([4,5,6,7], 2)
```

```python
df.CLASIFFICATION_FINAL.value_counts()
```

```python
df.head()
```

```python
plt.figure(figsize = (10,8))
ax = sns.countplot(df.CLASIFFICATION_FINAL)
for bars in ax.containers:
  ax.bar_label(bars)
plt.title("Count of covid and not covid", fontsize=25)
```

```python
#Split month and year in die date
df['MONTH'] = df.DATE_DIED.str.split("/" , expand=True)[1]
df['YEAR'] = df.DATE_DIED.str.split("/" , expand=True)[2]
```

```python
df.head()
```

```python
plt.figure(figsize = (10,8))
ax = sns.countplot(x = "YEAR", hue = "PATIENT_TYPE", data = df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Dead numbers of covid by year", fontsize = 25)
```

```python
plt.figure(figsize = (8,5))
ax = sns.countplot(x = "MONTH", hue = "PATIENT_TYPE", data = df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Dead numbers of covid by month", fontsize = 20)
```

```python
df.drop(["MONTH", "YEAR"], axis = 1, inplace = True)
df["CLASIFFICATION_FINAL"] = df["CLASIFFICATION_FINAL"] - 1
df.drop(["DATE_DIED"], axis = 1, inplace = True)
df.head()
```

```python
df.CLASIFFICATION_FINAL.value_counts()
```

#XGBOOST MODEL

```python
x = df.drop("CLASIFFICATION_FINAL", axis = 1)
y = df["CLASIFFICATION_FINAL"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(
 learning_rate =0.1,
 n_estimators=1000,
 max_depth=9,
 min_child_weight=7,
 gamma=0.1,
 subsample=1,
 colsample_bytree=0.4,
 objective= 'binary:logistic',
 nthread=4,
 scale_pos_weight=1,
 seed=27)

xgb.fit(x_train, y_train)

pred = xgb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, roc_curve, roc_auc_score, confusion_matrix, classification_report, plot_confusion_matrix, precision_score, recall_score, f1_score
accuracy_score(pred, y_test)
roc_auc_score(pred, y_test)

print("XGB Model's Accuracy is "+str(accuracy_score(pred, y_test)))
print("XGB Model's AUC score is "+str(roc_auc_score(pred, y_test)))
```
